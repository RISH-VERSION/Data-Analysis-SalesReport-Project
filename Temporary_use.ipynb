```python
import pandas as pd
```

```python
data = {
    "Id" : [1, 2, 3, 4, 5],
    "Name": ["John", "Alice", "Hari", "Tom", "Bob"],
    "Salary": [1000, 1300, 1500, 1250, 2000]
}


df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.heatmap(df[["Id", "Salary"]])
plt.figure()
sns.heatmap(df[["Salary", "Id"]])
# sns.barplot(x='Id', y='Salary', data=df)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
sns.heatmap(data=df[["Id", "Salary"]], ax=axes[0])
sns.heatmap(data=df[["Salary", "Id"]], ax=axes[1])

plt.show()
```
