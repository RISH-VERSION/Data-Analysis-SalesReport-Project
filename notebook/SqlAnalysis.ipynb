# SQL Analysis

```python
# Top selling products
top_selling_products_query = """select `Product Name` as Top_selling_products from superstore 
           group by `Product Name` 
           order by sum(Quantity) desc
           limit 10"""
pd.read_sql(top_selling_products_query, con=engine)
```

```python
# Highest Revenue Categories
highest_revenue_categories_query = """ select Category As Highest_Revenue_Categories from superstore
                                       group by Category 
                                       order by sum(Sales) desc"""
pd.read_sql(highest_revenue_categories_query, con=engine)
```

```python
# Monthly Sales Report
monthly_sales_report_query = """ select `Order Month`, sum(Sales) as Monthly_sales_report from superstore
                                 group by `Order Month`  order by `Order Month`"""

pd.read_sql(monthly_sales_report_query, con=engine)
```

```python
# Customer Analysis 
customer_analysis_with_order_counts_query = """ select `Customer Name`, sum(`Order Count`) as Total_Orders from superstore
                                                group by `Customer Name` order by sum(`Order Count`) desc 
                                                limit 10"""

pd.read_sql(customer_analysis_with_order_counts_query, con=engine)
```

```python
customer_analysis_total_spend_query = """ select `Customer Name`, sum(Sales) as Total_Spend from superstore
                                          group by `Customer Name` order by sum(Sales) desc 
                                          limit 10"""

pd.read_sql(customer_analysis_total_spend_query, con=engine)
```

```python
# Regional Sales Comparison
regional_sales_comparison_query = """ select Region, sum(Sales) as Total_Sales from superstore
                                      group by Region order by sum(sales) desc"""
pd.read_sql(regional_sales_comparison_query, con=engine)
```

```python
# Top Profitable Products
top_profitable_products_query = """ select `Product Name` as Top_Profitable_Products, sum(Profit) as Total_Profits from superstore
                                    group by `Product Name` order by sum(Profit) desc
                                    limit 10"""
pd.read_sql(top_profitable_products_query, con=engine)
```

```python
# Customer Segmentation
customer_segment_analysis_query = """select Segment, count(*) as Total_Orders, sum(Sales) as Total_Sales, sum(Profit) as Total_Profit from superstore
                                     group by Segment order by count(*) desc, sum(Sales) desc, sum(Profit) desc
                                  """

pd.read_sql(customer_segment_analysis_query, con=engine)
```

```python
# Sales Trend
sales_trend_query = """ select `Order Year` as Years, `Order Month` as Months, sum(Sales) as Total_Sales from superstore
                        group by `Order Year`, `Order Month` order by `Order Year`, `Order Month`"""

pd.read_sql(sales_trend_query, con=engine)
```
