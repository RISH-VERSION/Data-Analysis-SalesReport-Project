```python
import pandas as pd
```

```python
#Ensure the correct encoding is used to read the CSV file
df = pd.read_csv('../data/Raw-Superstore.csv', encoding='cp1252')
```

```python
#Analyze datatypes
print(df.dtypes)
```

```python
#Analyze numerical decriptive statistics
print(df.describe(include='number'))
```

```python
#Analyze categorical decriptive statistics
print(df.describe(include='str'))
```

```python
#Identify missing values
missing_values = df.isnull()
print(missing_values)
```

```python
#Calculating missing values
print(missing_values.sum())
```

```python
#Analyze the duplicated rows
duplicated_values = df.duplicated()
print(duplicated_values)
```

```python
#Calculating the duplicated values
print(duplicated_values.sum())
```

```python
# Data Visualizing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#heatmap needs 2d matrix so reusing missing_values 
data = missing_values
sns.heatmap(data)
plt.title("Missing_values") # x-axis and y-axis names extract from data(dataframe) object.
plt.show()
```

```python
# Feature Distribution using Histogram
# Profit Histogram
plt.hist(df['Profit'], bins=50)
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sales Histogram
plt.hist(df['Sales'], bins=50)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Quantity Histogram
plt.hist(df['Quantity'], bins=50)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Discount Histogram
plt.hist(df['Discount'], bins=10)
plt.title('Discount Distribution')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```
